## Terms
-weight - m value
cost - error(diff b/w actual and predicted value):-also called lost function
global minima - minimum VALUE
intercept and slope :- parameter value
step learning rate:- Derivative of cost:- learning step
gradient descent

```python
def gradient_descent(x,y):
    global all_cost
    all_cost =[]
    m = c = 0
    n = len(x)
    rate = 0.01
    for i in range(1000):
        
        y_pred = m*x + c
        plt.plot(x,y_pred)
        #print(i,m,c)
        md = (-2/n)*sum(x*(y-y_pred))
        yd = (-2/n)*sum(y-y_pred)
        m = m - rate*md
        c = c - rate*yd
        cost = (1/n)*sum((y-y_pred)**2)
        all_cost.append(cost)

        print(f'i:{i},m:{m},c:{c},cost:{cost}')
   # return all_cost

import numpy as np
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
        
```

```python
gradient_descent(x,y)
```
